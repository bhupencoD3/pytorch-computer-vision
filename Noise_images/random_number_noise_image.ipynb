```python
import torch
```

```python
random_tensor = torch.rand(5)
print(random_tensor)
print(random_tensor.ndim)
```

```python
random_tensor = torch.rand(2,5,3)
print(random_tensor)
print(random_tensor.ndim)
print(random_tensor.shape)
```

```python
torch.manual_seed(42) # for creating tensor with same values at each iteration
random_tensor = torch.rand(5)
print(random_tensor)
```

### generate random integers

```python
random_integers = torch.randint(low=-20,high=20,size=(5,5))
print(random_integers)
```

```python
## generate random number from normal dist

normal_tensor = torch.normal(mean=0.5,std=2,size=(5,))
print(normal_tensor)
```

### Creating Noise image using PyTorch

```python
from matplotlib import pyplot as plt
# !conda install -c conda-forge matplotlib -y
```

```python
# h,w of image(256x256)

uniform_noise = torch.rand((256,256))

plt.imshow(uniform_noise.numpy(),cmap='grey')
plt.title("Uniform Noise Image")
plt.axis('off')
plt.show()
```

```python
uniform_noise_rgb = torch.rand((256,256,3))

plt.imshow(uniform_noise_rgb.numpy())
plt.title("Uniform Noise RGB Image")
plt.axis('off')
plt.show()
```

```python
print("shape:",uniform_noise_rgb.shape)
print("DataType:",uniform_noise_rgb.dtype)
print("Min Value:",torch.min(uniform_noise_rgb))
print("Max Value:",torch.max(uniform_noise_rgb))
```
